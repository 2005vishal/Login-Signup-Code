```python
import pandas as pd
from datetime import datetime

datafr = pd.DataFrame(columns=['Name', 'LogIn_ID', 'Password', 'sequrity_key1', 'sequrity_key2'])
data_admin = pd.DataFrame({
    'Name': ['admin'],
    'LogIn_ID': ['demo'],
    'Password': ['demo123@'],
    'signUp_date' : ['05-01-2024']
})

# this funcation is check the password strenght
def checkPasswordStrenght():
  while True:
      password = input("Enter the Strong Password: ")
      s = ("[@_!#$%^&*()<>?/|}{~:]")
      uc=lc=ac=dc=sc=t=0
      if (len(password))>=8:
              for i in password:
                       if i.isalpha():
                              ac+=1
                              if i.isupper(): uc+=1
                              else: lc+=1
                       elif i.isdigit(): dc+=1
                       elif i in s : sc+=1
              if len(password)>=8: 
                    t+=1
                    if ac>0 and dc>0: t+=1
                    if uc>0 and lc>0: t+=1
                    if sc>0: t+=1
                    print("Password Strenght:",t)
                    return 'successful'
      else:
              print("password should be minimum 8 characters")

    

# this funcation is registered the user in database
def signUp():
    global datafr
    current_datetime = datetime.now()
    formatted_date = current_datetime.strftime("%d-%m-%Y")
    print('***Welcome to signUp Portal***')
    
    while True:
        Name = input('Enter Your Name or Press 0 to Exit: ').title()
        if Name == '0':
            return "Failed"
        else:
            userUp = input('''Enter the new Username: ''').lower()
            if not any(datafr['LogIn_ID'] == userUp):
                detail = checkPasswordStrenght()
                # Add your password strength checking logic here
                # checkPasswordStrenght(passUp)
                question1 = input('Your First School Name: ').upper()
                question2 = input('Your Best Friend Name: ').upper()
                    
                user_signin_detail = pd.DataFrame({
                    'Name': Name,
                    'SignUp_date': formatted_date,
                    'LogIn_ID': userUp,
                    'Password': detail,
                    'sequrity_key1': question1,
                    'sequrity_key2': question2
                }, index=[0])
                        
                datafr = pd.concat([datafr, user_signin_detail], ignore_index=True)
                print(datafr)
                return 'Successful' 
            else:
                print("Username already exists. Please try again....")

# signUp()


# this funcatio in forgot password 
def forgotPassword(): 
    userId=input("Enter your Userid: ").lower()
    if all(datafr['LogIn_ID'] == userId): 
            securityquestion1=input("Your First School Name: ").upper()
            securityquestion2=input('Your Best Friend Name: ').upper()
            if all(datafr['sequrity_key1'] == securityquestion1) and all(datafr['sequrity_key2'] == securityquestion2):
                    newpass=input("Enter your New Password: ")
                    datafr['Password'] = newpass
                    print('Successful Password Change')
                    return 
            else: print("Data Not Found")
    else: print("UserId Invalid")

# forgotPassword()

# this funcation Admin login funcatio
def admin_signIn():
    print('***Welcome to signIn Portal***')
    userIn = input('''Enter the Username
    or Press 0 to Exit: ''').lower()
    if userIn == '0':
            return 'Failed'
    elif 'LogIn_ID' in data_admin.columns:
            specific_ID = data_admin.loc[data_admin['LogIn_ID'] == userIn]
            if not specific_ID.empty:
                for i in range(3):
                     PassIn = input('Enter Your Password: ')
                     if any(specific_ID['Password'] == PassIn):
                         return ('Successfully Admin Login')
                     else:
                         print('Passward if Wrong.Please try again.... ')
                         print(f'Chances left{3-i-1}')
                con = input("Forgot password: y/n: ").lower() 
                if con.lower()== 'y': 
                    forgotPassword()
                    return 
                else: print('Failed')
            else: print('Failed')
    else:print('Failed')


# normal user login funcatio
def signIn():
    print('***Welcome to signIn Portal***')
    while 1:
        userIn = input('''Enter the Username
        or Press 0 to Exit: ''').lower()
        if userIn == '0':
            return 'Failed'
        elif 'LogIn_ID' in datafr.columns:
            specific_ID = datafr.loc[datafr['LogIn_ID'] == userIn]
            if not specific_ID.empty:
                for i in range(3):
                     PassIn = input('Enter Your Password: ')
                     if any(specific_ID['Password'] == PassIn):
                         return ('Successfully Login')
                     else:
                         print('Passward if Wrong.Please try again.... ')
                         print(f'Chances left{3-i-1}')
                con = input("Forgot password: y/n: ").lower() 
                if con.lower()== 'y': 
                    forgotPassword()
                    return 
                else: print('Failed')
            else: print('Failed')
        else:print('Failed')

#this is main code complete funcation in control
print("....welcome coding Thinker....")
while True:
    mode=input("""
    Press 1 to SignUp
    Press 2 to SignIn
    press 3 for admin Login
    press 0 to Exit 
    :""")
    if mode=="1":
        print('SignUP',signUp())
    elif mode=='2':
        print('SignIn',signIn())
    elif mode=='3':
        print('SignIn',admin_signIn())
    elif mode=='0':
        print('Thank You. Please visit Again')
        break

```
